```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([p for p in 2*np.random.rand(10000,2)-1 if np.linalg.norm(p) < 1])
```

```python
def labelPoint(p):
    c = [p[0] > 0, p[1] > 0, np.sum(np.abs(p)) > 1]
    return sum([int(c[i])*2**i for i in range(3)])   
```

```python
y = [labelPoint(p) for p in data]
colors = {0:'black', 1:'red', 2:'blue', 3:'green', 4:'orange', 5:'purple', 6:'hotpink', 7:'yellow'}
```

```python
plt.figure(figsize=(5,5))
plt.scatter(data[:,0], data[:,1], s = 3, c=[colors[yp] for yp in y])
plt.axis('equal')
plt.show()
```
